1. Desarrolla un algoritmo que le pida al usuario por su nombre y que lo salude con su nombre, excepto si se llama “Ana”, “Juan”, y “Pedro”.

```python
Name = input("Digite su nombre por favor: ") #Le pedimos el nombre al usuario

#establecemos los condicionales de los nombres los cuales no debemos saludar con su nombre, saludando con un simple "Hola, como estas?"
if Name == "Ana" or Name == "ana" or Name == "ANA":
    print("Hola! como estas?")

elif Name == "Juan" or Name == "juan" or Name == "JUAN":
    print("Hola! como estas?")
  
elif Name == "Pedro" or Name == "pedro" or Name == "PEDRO":
    print("Hola! como estas?")

#Si ninguna condicional anterior se cumple, saludamos al usuario por su nombre
else:
    print("Hola" + " " + Name + "! " + "Como estas?")
```

2. Desarrolla un algoritmo que por medio de una función, le pida al usuario una lista de N elementos y como resultado invierta el sentido de la lista. 

```python
n =  int(input("Digite el numero de elementos de su lista: ")) #pedimos la cantidad de elementos
lista = []

for x in range(n):
    elemento = input(f"Digite elemento {x + 1}: ")
    lista.append(elemento) #el termino ".append" se encarga de concatenar los elementos de la lista

lista.reverse() #revertimos la lista obtenida

#imprimimos lo obtenido
print("Su lista invertida es: ", end=' ')
print(lista)
```

3. Desarrolla un algoritmo que le pregunte al usuario cuántos números de la serie de Fibonnaci quiere generar y que al final los imprima. 

```python
Fibonnaci = int(input("Digita la cantidad de terminos de la serie de Fibonnaci: "))

if Fibonnaci <= 0: #tiene que ser mayor que cero forzosamente
  print ("Intente de nuevo, avor de digitar numeros mayores a cero")

else:
  print("imprimiendo serie de Fibonnaci...")
  s, t = 0, 1 #inicio de la serie, primeros 2 numeros

  for x in range(Fibonnaci):
    print(s)
    s, t = t, s + t #formula para la obtención de los numeros de fibonnaci
```
